```python
from ultralytics import YOLO

from glob import glob
from tqdm import tqdm
```

```python
MODEL_PATH = "models/yolov8/best2.pt"

IMAGES_PATH = "datasets/raw/images3"
LABELS_PATH = "datasets/raw/labels3"
```

```python
model = YOLO(MODEL_PATH)
```

```python
for image in tqdm(glob(f"{IMAGES_PATH}/*.jpg")):
    model(image, save_txt=True)
```
